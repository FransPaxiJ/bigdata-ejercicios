```python
# Creamos un dataframe 'data_peru' que contendrÃ¡ los registros de PerÃº
data_peru = data[data['location'] == 'Peru']
```

```python
# Regstramo y creamos el dataframe con 3 columnas
data_peru = data_peru[['location', 'date', 'new_cases_smoothed']]
data_peru
```

```python
# Nos piden los datos de agosto 2020 en Peru
# Para ello convertimos la columna 'date' a formato propio 'Datetime' de pandas
# Para realizar operaciones de filtrado en dicha columna
data_peru['date'] = pd.to_datetime(data_peru['date'])
```

```python
data_peru
```

```python
# Filtramos los datos de PerÃº en Agosto 2020 a partir del dataframe 'data_peru' en un 
# nuevo dataframe 'data_peru_agosto_2020'
data_peru_agosto_2020 = data_peru[(data_peru["date"].dt.year == 2020) & (data_peru["date"].dt.month == 8)]
pd.set_option('display.max_rows', None)
data_peru_agosto_2020
```

```python
# Mostrar en grafico de barras en linea
data_peru_agosto_2020.plot(x='date', y='new_cases_smoothed', kind='bar')
```

```python
# Filtrar datos de los paises andinos (Peru, Argentina, Bolivia y Chile)
paises_andinos = data[(data['location'] == 'Peru') | (data['location'] == 'Argentina') | (data['location'] == 'Bolivia') | (data['location'] == 'Chile')]
paises_andinos
```

```python
paises_andinos = paises_andinos[['location', 'date', 'new_cases_smoothed']]
paises_andinos
```

```python
# Dado que en la columna new_cases_smoothed hay valores NaN, se procede a reemplazarlos por 0
# o es mejor eliminar los registros con valores NaN

# primer eliminamos los registros con valores NaN y los 0
paises_andinos_nn = paises_andinos.dropna(subset=['new_cases_smoothed'])
paises_andinos_nn
```

```python
# Agrupamos por pais y sumamos la columna new_cases_smoothed y tener en cuenta que el resultado
# sera un objeto de tipo 'Series' y no un dataframe
paises_andinos_nn = paises_andinos_nn.groupby('location')['new_cases_smoothed'].sum()

# Convertimos el objeto 'Series' a un dataframe
pd.set_option('display.float_format', lambda x: '%.0f' % x) # Formateamos los numeros flotantes
paises_andinos_nn
```

```python
# trabajamos con paises_andinos_mes para no estropear el dataframe paises_andinos
paises_andinos_mes = paises_andinos.dropna(subset=['new_cases_smoothed'])
paises_andinos_mes
```

```python
# Para ello convertimos la columna 'date' a formato propio 'Datetime' de pandas
# Para realizar operaciones de filtrado en dicha columna
paises_andinos_mes['date'] = pd.to_datetime(paises_andinos_mes['date'])
paises_andinos_mes = paises_andinos_mes.groupby([paises_andinos_mes['date'].dt.year, 'location'])['new_cases_smoothed'].sum().reset_index()
# Agregamos una columna 'total' que es la suma de 'new_cases_smoothed' de todos los aÃ±os 2020 hasta el 2024
paises_andinos_mes['total'] = paises_andinos_mes.groupby('location')['new_cases_smoothed'].transform('sum')

# Convertimos el objeto 'Series' a un dataframe
pd.set_option('display.float_format', lambda x: '%.0f' % x) # Formateamos los numeros flotantes
paises_andinos_nn
```
